```python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import telebot
from tensorflow.keras.models import load_model


# Download NLTK resources
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')

# Preprocessing function
def preprocess_text(text):
    # Convert text to lowercase
    text = text.lower()

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Tokenize text into individual words
    tokens = word_tokenize(text)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]

    # Lemmatize words
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]

    # Join the tokens back into a single string
    processed_text = ' '.join(tokens)

    return processed_text

# Load the toxic comment detection model
model = load_model("C:\\Users\\amaoa\\Downloads\\toxicity.h5")

# Initialize the Telegram bot
bot = telebot.TeleBot("6093147054:AAFqMEhR0ReR2Jk4MOr_gnZfjqnv_URkTSY")

# Predefined threshold for classifying as toxic or non-toxic
threshold = 0.5

# Handle incoming messages
@bot.message_handler(func=lambda message: True)
def handle_message(message):
    # Preprocess user input
    processed_text = preprocess_text(message.text)

    # Make prediction using the toxic comment detection model
    prediction = model.predict([processed_text])[0][0]

    if prediction >= threshold:
        # Comment is toxic
        response = "Your comment is flagged as toxic. Please refrain from using offensive language."
    else:
        # Comment is non-toxic
        response = "Thank you for your comment. We appreciate your contribution."

    # Send the response back to the user
    bot.reply_to(message, response)

# Start the bot
bot.polling()

bot.send_message("Hello, bot!")
bot.send_message("This is a friendly message.")
bot.send_message("I really like your service!")
bot.send_message("You guys are doing a great job.")
bot.send_message("You're the worst!")
bot.send_message("I can't believe how terrible your product is.")
bot.send_message("Do you have any special offers?")
bot.send_message("What are your customer support hours?")
bot.send_message("Can you help me with a technical issue?")
bot.send_message("I want to cancel my subscription.")
```
